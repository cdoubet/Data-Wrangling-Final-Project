# Analysis

# Question 1: Who are the top ten Billionaires in terms of wealth? What countries do they reside in? What industry did they obtained their wealth in?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

billionaires['net_worth'] = billionaires['net_worth'].astype(str).str.replace(' B', '').astype(float)

# Find the most recent years
latest_year = billionaires['year'].max()
latest_billionaires = billionaires[billionaires['year'] == latest_year]

# Sort and get top 10 billionaires
top_10_billionaires = latest_billionaires.sort_values('net_worth', ascending=False).head(10)

# Create the datafarme of top ten billionaires
top_df = top_10_billionaires[['full_name', 'country_of_residence', 'business_industries', 'net_worth']]

print("Top 10 Billionaires:")
print(top_df.to_string(index=False))

# Visualization
plt.figure(figsize=(12, 6))
plt.bar(top_10_billionaires['full_name'], top_10_billionaires['net_worth'])
plt.title(f'Top 10 Billionaires Wealth in {latest_year}')
plt.xlabel('Billionaire Name')
plt.ylabel('Net Worth (Billions USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Country distribution
country_distribution = top_10_billionaires['country_of_residence'].value_counts()

plt.figure(figsize=(10, 6))
country_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Country Distribution of Top 10 Billionaires')
plt.ylabel('')
plt.tight_layout()
plt.show()

print("\nCountry Distribution of Top 10 Billionaires:")
print(country_distribution)
```

```python
# Industry distribution
def extract_primary_industry(industries):
    if isinstance(industries, list):
        return industries[0] if industries else 'Unknown'
    elif isinstance(industries, str):
        import ast
        try:
            industry_list = ast.literal_eval(industries)
            return industry_list[0] if industry_list else 'Unknown'
        except:
            return industries
    return 'Unknown'

top_10_billionaires['primary_industry'] = top_10_billionaires['business_industries'].apply(extract_primary_industry)
industry_distribution = top_10_billionaires['primary_industry'].value_counts()

plt.figure(figsize=(10, 6))
industry_distribution.plot(kind='bar')
plt.title('Primary Industries of Top 10 Billionaires')
plt.xlabel('Industry')
plt.ylabel('Number of Billionaires')
plt.tight_layout()
plt.show()

print("\nIndustry Distribution of Top 10 Billionaires:")
print(industry_distribution)
```

# Question 2: Do a lot of Billionaires symbolize a stable economy and lead to steady GDP for a country? Does few Billionaires symbol volatile GDP?

```python
# Group by country and calculate key metrics
country_billionaire = billionaires_gdp.groupby('country_of_residence').agg({
    'full_name': 'count',  
    '2023_gdp_change': 'first',  
    '2022_gdp_change': 'first',
    '2021_gdp_change': 'first',
    '2020_gdp_change': 'first',
    '2019_gdp_change': 'first'
}).reset_index()

# Rename columns for clarity
country_billionaire.columns = ['Country', 'Billionaire_Count', 'GDP_2023', 'GDP_2022', 'GDP_2021', 'GDP_2020', 'GDP_2019']

# Calculate GDP volatility (standard deviation of GDP changes)
country_billionaire_analysis['GDP_Volatility'] = country_billionaire_analysis[
    ['GDP_2023', 'GDP_2022', 'GDP_2021', 'GDP_2020', 'GDP_2019']
].std(axis=1)

# Calculate correlation between billionaire count and GDP volatility
correlation = country_billionaire_analysis['Billionaire_Count'].corr(country_billionaire_analysis['GDP_Volatility'])
```

```python
# Visualize the relationship
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.scatterplot(
    data=country_billionaire_analysis, 
    x='Billionaire_Count', 
    y='GDP_Volatility',
    alpha=0.7
)
plt.title('Billionaire Count vs GDP Volatility')
plt.xlabel('Number of Billionaires')
plt.ylabel('GDP Volatility (Standard Deviation)')
plt.show()

# Print key findings
print("Correlation between Billionaire Count and GDP Volatility:", correlation)

# Top 10 countries by billionaire count and their GDP volatility
top_billionaire_countries = country_billionaire_analysis.sort_values('Billionaire_Count', ascending=False).head(20)
print("\nTop 10 Countries by Billionaire Count:")
print(top_billionaire_countries[['Country', 'Billionaire_Count', 'GDP_Volatility']])
```

# Question 3: Do countries with a lot of billionaires have low unemployment rates signaling healthy economies? Or do countries with a lot of billionaires have high unemployment rates due to the concentration of wealth?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

billionaires_unemployment = pd.read_csv("Billionaires_Unemployment.csv", sep=',', encoding='utf-8')

billionaires_per_country = billionaires_unemployment.groupby('country_of_residence').size().reset_index(name='billionaire_count')

unemployment_columns = [
    '2019_unemployment_rate', 
    '2020_unemployment_rate', 
    '2021_unemployment_rate', 
    '2022_unemployment_rate', 
    '2023_unemployment_rate'
]

avg_unemployment = billionaires_unemployment.groupby('country_of_residence')[unemployment_columns].mean().reset_index()
avg_unemployment['average_unemployment_rate'] = avg_unemployment[unemployment_columns].mean(axis=1)

merged_data = pd.merge(billionaires_per_country, avg_unemployment, on='country_of_residence')

plt.figure(figsize=(12, 6))
plt.scatter(merged_data['billionaire_count'], merged_data['average_unemployment_rate'], alpha=0.6)
plt.title('Billionaires Count vs Average Unemployment Rate')
plt.xlabel('Number of Billionaires')
plt.ylabel('Average Unemployment Rate (%)')

plt.show()

correlation = merged_data['billionaire_count'].corr(merged_data['average_unemployment_rate'])
print(f"Correlation between Billionaire Count and Unemployment Rate: {correlation}")

top_10_countries = merged_data.sort_values('billionaire_count', ascending=False).head(20)
print("\nTop 20 Countries by Billionaire Count and Their Unemployment Rates:")
print(top_10_countries[['country_of_residence', 'billionaire_count', 'average_unemployment_rate']])
```
